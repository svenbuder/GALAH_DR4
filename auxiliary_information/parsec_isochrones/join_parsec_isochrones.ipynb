```python
for fe_h in fe_h_array:
    fe_h_string = "{:.2f}".format(abs(fe_h)).replace('.','p')
    if fe_h < 0:
        fe_h_string = 'm'+fe_h_string

    Gaia_isochrone_one_metallicity = np.genfromtxt('GaiaEDR3/GaiaEDR3_age_6p19_0p01_10p18_m_h_'+fe_h_string+'.dat')

    m_h.append(Gaia_isochrone_one_metallicity[:,1])
    logAge.append(Gaia_isochrone_one_metallicity[:,2])
    mass.append(Gaia_isochrone_one_metallicity[:,5])
    logL.append(Gaia_isochrone_one_metallicity[:,6])
    logT.append(Gaia_isochrone_one_metallicity[:,7])
    logg.append(Gaia_isochrone_one_metallicity[:,8])
    Gmag.append(Gaia_isochrone_one_metallicity[:,28])
    GBPmag.append(Gaia_isochrone_one_metallicity[:,29])
    GRPmag.append(Gaia_isochrone_one_metallicity[:,30])
    # 0 Zini     
    # 1 MH   
    # 2 logAge 
    # 3 Mini        
    # 4 int_IMF         
    # 5 Mass   
    # 6 logL    
    # 7 logTe  
    # 8 logg  
    # 9 label   
    # 10 McoreTP 
    # 11 C_O  
    # 12 period0  
    # 13 period1  
    # 14 period2  
    # 15 period3  
    # 16 period4  
    # 17 pmode  
    # 18 Mloss  
    # 19 tau1m   
    # 20 X   
    # 21 Y   
    # 22 Xc  
    # 23 Xn  
    # 24 Xo  
    # 25 Cexcess  
    # 26 Z 	 
    # 27 mbolmag  
    # 28 Gmag    
    # 29 G_BPmag  
    # 30 G_RPmag

    TwoMASS_isochrone_one_metallicity = np.genfromtxt('2MASS_WISE/2MASS_WISE_age_6p19_0p01_10p18_m_h_'+fe_h_string+'.dat')

    Jmag.append(Gaia_isochrone_one_metallicity[:,28])
    Hmag.append(Gaia_isochrone_one_metallicity[:,29])
    Ksmag.append(Gaia_isochrone_one_metallicity[:,30])
    # W1mag.append(Gaia_isochrone_one_metallicity[:,38])
    # W2mag.append(Gaia_isochrone_one_metallicity[:,39])
    # W3mag.append(Gaia_isochrone_one_metallicity[:,40])
    # W4mag.append(Gaia_isochrone_one_metallicity[:,41])
    # 0 Zini
    # 1 MH   
    # 2 logAge 
    # 3 Mini        
    # 4 int_IMF         
    # 5 Mass   
    # 6 logL    
    # 7 logTe  
    # 8 logg  
    # 9 label   
    # 10 McoreTP 
    # 11 C_O  
    # 12 period0  
    # 13 period1  
    # 14 period2  
    # 15 period3  
    # 16 period4  
    # 17 pmode  
    # 18 Mloss  
    # 19 tau1m
    # 20 X   
    # 21 Y   
    # 22 Xc  
    # 23 Xn  
    # 24 Xo  
    # 25 Cexcess  
    # 26 Z 	 
    # 27 mbolmag  
    # 28 Jmag    
    # 29 Hmag  
    # 30 Ksmag 
    # 31 IRAC_3.6mag
    # 32 IRAC_4.5mag  
    # 33 IRAC_5.8mag  
    # 34 IRAC_8.0mag  
    # 35 MIPS_24mag  
    # 36 MIPS_70mag  
    # 37 MIPS_160mag  
    # 38 W1mag   
    # 39 W2mag   
    # 40 W3mag   
    # 41 W4mag
```

```python
isochrones = Table()
isochrones['m_h'] = np.concatenate(m_h)
isochrones['logAge'] = np.concatenate(logAge)
isochrones['mass'] = np.concatenate(mass)
isochrones['logL'] = np.concatenate(logL)
isochrones['logT'] = np.concatenate(logT)
isochrones['logg'] = np.concatenate(logg)
isochrones['Gmag'] = np.concatenate(Gmag)
isochrones['GBPmag'] = np.concatenate(GBPmag)
isochrones['GRPmag'] = np.concatenate(GRPmag)
isochrones['Jmag'] = np.concatenate(Jmag)
isochrones['Hmag'] = np.concatenate(Hmag)
isochrones['Ksmag'] = np.concatenate(Ksmag)
# isochrones['W1mag'] = np.concatenate(W1mag)
# isochrones['W2mag'] = np.concatenate(W2mag)
# isochrones['W3mag'] = np.concatenate(W3mag)
# isochrones['W4mag'] = np.concatenate(W4mag)
```

```python
# get rid of some unnesessary entries:
restricted = (
    (isochrones['logT'] <= np.log(9999)) & # Teff < 10,000 K
    (isochrones['logg'] < 6) # Stick with main sequence or more luminous
)
restricted_isochrones = isochrones[restricted]
```
