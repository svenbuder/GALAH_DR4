```python
meta = []

version = '240705'

for catalogue in [
    'galah_dr4_allstar',
    'galah_dr4_allspec',
    'galah_dr4_vac_wise_tmass_gaiadr3',
    'galah_dr4_vac_dynamics',
    'galah_dr4_vac_3dnlte_a_li',
]:

    data,header = fits.getdata(catalogue+'_'+version+'.fits', header=True, ext=1)

    if catalogue == 'galah_dr4_allstar':
        table_name = 'mainstartable_v'+version
    elif catalogue == 'galah_dr4_allspec':
        table_name = 'mainspectable_v'+version
    elif catalogue == 'galah_dr4_vac_wise_tmass_gaiadr3':
        table_name = 'vacxmatchtable_v'+version
    elif catalogue == 'galah_dr4_vac_dynamics':
        table_name = 'vacdynamicstable_v'+version
    elif catalogue == 'galah_dr4_vac_3dnlte_a_li':
        table_name = 'vac3dnltelitable_v'+version
    else:
        raise ValueError('catalogue not known: '+catalogue)

    for ind,name in enumerate(data.dtype.names):
        
        if header['TFORM'+str(ind+1)] in ['K','I','J']:
            data_type = 'integer'
        elif header['TFORM'+str(ind+1)] in ['L']:
            data_type = 'boolean'
        elif header['TFORM'+str(ind+1)] in ['E','D']:
            data_type = 'float'
        elif header['TFORM'+str(ind+1)] in ['3A','4A','5A','6A','7A','13A','17A','16A','19A','28A']:
            data_type = 'string'
        else:
            print(header['TFORM'+str(ind+1)])

        meta.append(name+'|'+table_name+'_'+version+'|'+header['TCOMM'+str(ind+1)]+'|'+header['TUCD'+str(ind+1)]+'|'+unit+'|'+data_type)
np.savetxt('galah_dr4_column_meta.txt',np.array(meta),fmt='%s')
```
